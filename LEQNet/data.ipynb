```python
import pandas as pd
import h5py
import numpy as np
import matplotlib.pyplot as plt

file_name = "merge_modified.hdf5"
csv_file = "new_merge_modified.csv"
modified_file_name = "train_merge_modified1.hdf5"
modified_file_csv = "train_merge_modified1.csv"


df = pd.read_csv(csv_file)

df1 = df[0:30000]
df2 = df[235425 : 305425]

df3 = df1.append(df2)
df3.to_csv(modified_file_csv, index=None)

ev_list = df3['trace_name'].to_list()

dtfl = h5py.File(file_name, 'r')
processed = h5py.File(modified_file_name, 'w')
grp = processed.create_group("data")

for c, evi in enumerate(ev_list):
    dataset = dtfl.get('data/'+str(evi)) 
    dst = grp.create_dataset(str(evi), data=dataset)
    
    for key, value in dataset.attrs.items():
        dst.attrs[key] = value
        
    if c % 1000 == 0:
        print(str(c) + " / " + str(len(ev_list)))

dtfl.close()
processed.close()
```

```python
import pandas as pd
import h5py
import numpy as np
import matplotlib.pyplot as plt
import random
import csv

file_name = "train_merge_modified1.hdf5"
csv_file = "train_merge_modified1.csv"
modified_file_name = "mix_train_merge_modified1.hdf5"
modified_file_csv = "mix_train_merge_modified1.csv"


df = pd.read_csv(csv_file, index_col=0)

num = list(range(100000))
random.shuffle(num)

new_df = df[0:1]

for c, i in enumerate(num):
    new_df = new_df.append(df[i:i+1])
    if c % 1000 == 0:
        print(str(c) + " / " + str(len(num)))

new_df=new_df[1:]
new_df.to_csv(modified_file_csv)

ev_list = new_df['trace_name'].to_list()

dtfl = h5py.File(file_name, 'r')
processed = h5py.File(modified_file_name, 'w')
grp = processed.create_group("data")

for c, evi in enumerate(ev_list):
    dataset = dtfl.get('data/'+str(evi)) 
    dst = grp.create_dataset(str(evi), data=dataset)
    
    for key, value in dataset.attrs.items():
        dst.attrs[key] = value
        
    if c % 1000 == 0:
        print(str(c) + " / " + str(len(ev_list)))

dtfl.close()
processed.close()
```
